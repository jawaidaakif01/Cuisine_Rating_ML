```python
df.columns = df.columns.str.strip()
df['Marital_Status_grouped'] = df['Marital Status'].replace({
    'Single':"Not_Married",
    "Divorced":"Not_Married"
})
```

```python
df['Budget_grouped'] = df['Budget'].replace({
    1:"3_andBelow3",
    2:"3_andBelow3",
    3:"3_andBelow3",
    4:"4",
    5:"5"
})
```

```python
df['Budget_grouped']
```

```python
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder
```

```python
ordinal_encoder = OrdinalEncoder(categories=[['Never', 'Socially', 'Often']])
df['Smoker_encoded'] = ordinal_encoder.fit_transform(df[['Smoker']])
df['Alcohol_encoded'] = ordinal_encoder.fit_transform(df[['Alcohol']])
```

```python
df.head()
```

```python
nominal_encoder = OneHotEncoder()
gender_encoded = nominal_encoder.fit_transform(df[['Gender']]).toarray()
encoded_df = pd.DataFrame(gender_encoded, columns=nominal_encoder.get_feature_names_out(['Gender']), index = df.index)
df = pd.concat([df, encoded_df], axis=1)
df.head()
```

```python
df['Gender'].value_counts()
```

```python
df['Gender_Female'].value_counts()
```

```python
df['Activity'].value_counts()
```

```python
encoder = OneHotEncoder()
encoded_activity = encoder.fit_transform(df[['Activity']]).toarray()
df_encoded = pd.DataFrame(encoded_activity, columns=encoder.get_feature_names_out(['Activity']), index=df.index)
df = pd.concat([df, df_encoded], axis=1)
df.head()
```

```python
df['Cuisines'].value_counts()
```

```python
encoder = OneHotEncoder()
location_encoded = encoder.fit_transform(df[['Location']]).toarray()
df_encoded = pd.DataFrame(location_encoded, columns=encoder.get_feature_names_out(['Location']), index=df.index)
df = pd.concat([df, df_encoded], axis=1)
df.head()
```

```python
encoder = OneHotEncoder()
Cuisines_encoded = encoder.fit_transform(df[['Cuisines']]).toarray()
df_encoded = pd.DataFrame(Cuisines_encoded, columns=encoder.get_feature_names_out(['Cuisines']), index = df.index)
df = pd.concat([df, df_encoded], axis = 1)
df.head()
```

```python
columns = ['Location', 'Gender', 'YOB', "Marital Status", 'Activity', 'Budget', 'Cuisines', 'Smoker', 'Alcohol']
df = df.drop(columns=['Location', 'Gender', 'YOB', "Marital Status", 'Activity', 'Budget', 'Cuisines', 'Smoker', 'Alcohol'])
df.head()
```
